## 1.2 Model representation

```python
# global parameters
input_layer_size = 400
hidden_layer_size = 25
output_layer_size = 10
```

```python
weight_mat = scipy.io.loadmat('data\ex4weights.mat')
theta1 = weight_mat['Theta1']
theta2 = weight_mat['Theta2']
theta = [theta1, theta2]

print(theta1.shape)
print(theta2.shape)
```

```python
# utility functions
def flattenParam(theta):
    tmp = [i.flatten() for i in theta]
    flt = list(itertools.chain.from_iterable(tmp))
    return np.array(flt).reshape((len(flt), 1))

def reshapeParam(theta):
    theta1 = theta[:hidden_layer_size * (input_layer_size + 1)].reshape((hidden_layer_size, input_layer_size + 1))
    theta2 = theta[hidden_layer_size * (input_layer_size + 1):].reshape((output_layer_size, hidden_layer_size + 1))
    return [theta1, theta2]

def flattenX(X):
    return np.array(X.flatten()).reshape((m * (input_layer_size + 1), 1))

def reshapeX(X):
    return np.array(X).reshape((m, input_layer_size + 1))
```

## 1.3 Feedforward and cost function

$ J(\theta)=-\frac{1}{m}[\sum_{i=1}^m y^{(i)} \log h_\theta(x^{(i)}) + (1 - y^{(i)})\log (1 - h_\theta(x^{(i)}))] + \frac{\lambda}{2m} \sum_{l=1}^{L-1} \sum_{i=1}^{s_l} \sum_{j=1}^{s_{l+1}} (\theta _{ji} ^{(l)})^2$

```python
# conduct foreward propagation
def foreProp(theta, X):
    x = X
    for i in range(len(theta)):
        itheta = theta[i]
        z = np.dot(itheta, x)
        a = expit(z) # activation function
        if i == len(theta) - 1:
            return a
        a = np.insert(a, 0, 1)
        x = a

# compute the overall cost function
def cost(theta_flattened, X_flattened, y, mylambda = 0.):
    total_cost = 0
    theta = reshapeParam(theta_flattened)
    X = reshapeX(X_flattened)
    
    # compute the total cost by adding up the cost of each training sample
    for i in range(m):
        x = X[i] # 401x1
        y_pred = foreProp(theta, x) # (10, 1)
        y_tmp = np.zeros((10, 1))
        y_tmp[y[i] - 1] = 1
        i_cost = - np.dot(y_tmp.T, np.log(y_pred)) - np.dot(1 - y_tmp.T, np.log(1 - y_pred))
        total_cost += i_cost
    
    total_cost = float(total_cost) / m
    
    total_reg = 0.
    for itheta in theta:
        total_reg += np.sum(itheta * itheta)
    total_reg *= float(mylambda) / (2 * m)
    
    return total_cost + total_reg
```

# 2 Backpropagation

## 2.1 Sigmoid gradient

```python
def sigmoidGrad(z):
    tmp = expit(z)
    return tmp * (1 - tmp)
```

```python
sigmoidGrad(0)
```

## 2.2 Random initialization

```python
def genRandTheta():
    epsilon_init = 0.12
    rand_theta1 = np.random.rand( * (hidden_layer_size, input_layer_size+1)) * 2 * epsilon_init - epsilon_init
    rand_theta2 = np.random.rand( * (output_layer_size, hidden_layer_size+1)) * 2 * epsilon_init - epsilon_init
    return [rand_theta1, rand_theta2]
```

## 2.3 Backpropagation

```python
# record z and a during the forward propagation
def forePropLog(theta, X):
    x = X
    log_z_a = []
    
    for i in range(len(theta)):
        
        itheta = theta[i]
        z = np.dot(itheta, x).reshape((itheta.shape[0], 1))
        a = expit(z) # activation function
        log_z_a.append((z, a))
        
        if i == len(theta) - 1:
            return np.array(log_z_a)
        
        a = np.insert(a, 0, 1)
        x = a
```

```python
def backProp(theta_flattened, X_flattened, y, mylambda = 0.):
    
    Delta1 = np.zeros(((hidden_layer_size, input_layer_size + 1)))
    Delta2 = np.zeros((output_layer_size, hidden_layer_size + 1))
    theta = reshapeParam(theta_flattened)
    X = reshapeX(X_flattened)
    
    for i in range(m):
        
        x = X[i] # (1, 401)
        a1 = x.reshape((input_layer_size + 1, 1))
        tmp = forePropLog(theta, x)
        
        z2 = tmp[0][0]
        a2 = tmp[0][1]
        z3 = tmp[1][0]
        a3 = tmp[1][1]
        
        y_tmp = np.zeros((10, 1))
        y_tmp[y[i] - 1] = 1
        
        delta3 = a3 - y_tmp
        delta2 = np.dot(theta[1].T[1:, :], delta3) * sigmoidGrad(z2)
        a2 = np.insert(a2, 0, 1, axis = 0)
        Delta1 += np.dot(delta2, a1.T) # (25, 1) x (1, 401)
        Delta2 += np.dot(delta3, a2.T) # (10, 1) x (1, 26)
        
    D1 = Delta1 / float(m)
    D2 = Delta2 / float(m)
    
    # regularization
    D1[:, 1:] = D1[:, 1:] + (float(mylambda) / m) * theta[0][:, 1:]
    D2[:, 1:] = D2[:, 1:] + (float(mylambda) / m) * theta[1][:, 1:]
    
    return flattenParam([D1, D2]).flatten()
```

```python
tmp = backProp(flattenParam(theta), flattenX(X), y, mylambda = 0.)
D1, D2 = reshapeParam(tmp)
```

## 2.4 Gradient checking

```python
def checkGrad(theta, D, X, y, mylambda = 0.):
    theta_flattened = flattenParam(theta)
    D_flattened = flattenParam(D)
    X_flattened = flattenX(X)
    n = len(theta_flattened)
    eps = 0.0001
    
    for i in range(10):
        x = int(np.random.rand() * n)
        epsvec = np.zeros((n, 1))
        epsvec[x] = eps
        
        cost_high = cost(theta_flattened + epsvec, X_flattened, y, mylambda)
        cost_low = cost(theta_flattened - epsvec, X_flattened, y, mylambda)
        grad = (cost_high - cost_low) / float(2 * eps)
        
        print("Element: {:d}. Numerical Gradient = {:f}. BackProp Gradient = {:f}".format(x, grad, float(D_flattened[x])))
```

```python
checkGrad(theta, [D1, D2], X, y)
```

## 2.6 Learning parameters using fmincg

```python
def trainNN(mylambda = 0.):
    theta_init_flattened = flattenParam(genRandTheta())
    result = scipy.optimize.fmin_cg(cost, x0 = theta_init_flattened, fprime = backProp, \
                                    args = (flattenX(X), y, mylambda), maxiter = 50, disp = True, full_output = True)
    return reshapeParam(result[0])
```

```python
Theta = trainNN()
```

```python
def pred(theta, x):
    cls = [i for i in range(1, 10)] + [10]
    output = forePropLog(theta, x)
    return cls[np.argmax(output[-1][1])]

def acc(theta, X, y):
    n_correct = 0
    for i in range(m):
        if int(pred(theta, X[i])) == int(y[i]):
            n_correct += 1
    
    print("Accuracy = {:%}".format(n_correct / m))
```

```python
acc(Theta, X, y)
```
